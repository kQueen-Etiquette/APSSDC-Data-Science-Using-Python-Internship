```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/AP-Skill-Development-Corporation/DataScienceUsingPython-Internship-SRM-University/main/Day12_Pandas/iris.csv")
df
```

```python
df.drop(['Unnamed: 0'], axis = 1)
```

```python
df.isnull().sum()
```

# Line Plot

```python
plt.plot(df['sepal length (cm)'], label = 'sepal length (cm)')
plt.plot(df['sepal width (cm)'], label = 'sepal width (cm)')
plt.legend()
plt.show()
```

# Scatter plot

```python
plt.scatter(df['petal length (cm)'], df['petal width (cm)'])
plt.xlabel('petal length (cm)')
plt.ylabel('petal width (cm)')
plt.show()
```

# Histogram

```python
plt.hist(df['sepal length (cm)'])
plt.xlabel('sepal length (cm)')
plt.show()
```

# Bar Graph

```python
plt.bar(df['sepal width (cm)'], df['petal length (cm)'], width = 0.5)
plt.xlabel('sepal width (cm)')
plt.ylabel('petal length (cm)')
plt.show()
```

# Box Plot

```python
plt.boxplot(df['sepal length (cm)'])
plt.boxplot(df['petal width (cm)'])
plt.show()
```

# Pie Chart

```python
plt.pie(df['sepal length (cm)'][:10], explode = [0, 0, 0, 0.6, 0, 0, 0, 0, 0, 0], shadow = True)
plt.show()
```

# Subplot

```python
fig, area = plt.subplots(2, 2)
area[0][0].plot(df['sepal length (cm)'])
area[0][1].plot(df['sepal width (cm)'])
area[1][0].plot(df['petal length (cm)'])
area[1][1].plot(df['petal width (cm)'])
plt.show()
```

```python
plt.subplot(1, 2, 1)
plt.plot(df['sepal length (cm)'])
plt.subplot(1, 2, 2)
plt.plot(df['petal length (cm)'])
plt.show()
```
